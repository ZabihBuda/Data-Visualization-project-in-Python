```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import random as random
pd.set_option('display.max_columns', None)
```

```python
#importing the dataset
df=pd.read_csv('agricultural_raw_material.csv')
df.head(40)
```

```python
#checking the missing values
df.isnull().sum()
```

```python
df.shape
```

```python
#dropping the null values
```

```python
df.head()
```

```python
# Replacing %, "," and "-"
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

df=df.dropna()
print(df.isnull().sum())
print(df.shape)
```

```python
# Converting data type to float
lst = ["Coarse wool Price", "Coarse wool price % Change", "Copra Price", "Copra price % Change", "Cotton price % Change","Fine wool Price", "Fine wool price % Change", "Hard log price % Change", "Hard sawnwood price % Change", "Hide price % change", "Plywood price % Change", "Rubber price % Change", "Softlog price % Change", "Soft sawnwood price % Change", "Wood pulp price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
#Datetime Columns- formating the datetime columns and setting it as the index
df['Month']=pd.to_datetime(df['Month'].str.upper(), format='%b%y', yearfirst=False)
#Indexing the Month
df=df.set_index('Month')
df.head()
```

# Data Visualization and Exploration

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

# Heatmap
raw_data=['Coarse wool Price',  'Copra Price','Cotton Price', 'Fine wool Price',  'Hard log Price', 'Hard sawnwood Price','Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']

#correlations
corrmat = df[raw_data].corr()
#plot size
fig=plt.figure(figsize=(12, 8))
#masking the upper triangle part since matrix is symmetrics(repetitive)
mask=np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax= .8, mask=mask, square=True, annot=True)
plt.show()
```

```python
#Plotting the prices and their % changes
axes=df[['Coarse wool Price', 'Coarse wool price % Change']].plot(figsize=(20, 9), subplots=True, linewidth=1)
```

```python
# finding out the normal price changes of each raw material
Collist= ['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']
for i in range(len(Collist)):
    plt.figure(figsize=(12,12))
    df[Collist[i]].hist(figsize=(11, 9), linewidth=1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(Collist[i:],loc='upper center',bbox_to_anchor=(1.2,1))
```

It seems like the most of raw materials ahve ideal frequent $ changes less thant 5%.

```python
# Finding the raw materials that has the lowest price over years
plt.figure(figsize=(30, 15))
list_materials=['Copra Price','Softlog Price','Rubber Price','Cotton Price','Coarse wool Price',
                'Fine wool Price','Hard log Price','Hard sawnwood Price','Hide Price','Plywood Price',
                'Soft sawnwood Price','Wood pulp Price']
for i in range(len(list_materials)):
  plt.subplot(4, 3, i+1)
  plt.subplots_adjust(hspace=1, wspace=0.5)
  plt.title(list_materials[i])
  plt.plot(df[list_materials[i]])
  plt.xticks(rotation=90)
plt.suptitle('Raw Materials Price Comparision over the years')
```

Rubber and Cotton have the lowest price

```python
#finding the raw materials which has the lowest and highest prices % changes
price_change=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change',
              'Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change',
              'Soft sawnwood price % Change','Wood pulp price % Change']
plt.figure(figsize=(20,20))
for i in range(len(price_change)):
  r=random.random()
  b=random.random()
  g=random.random()
  color=(r, b, g)
  plt.subplot(4,3,i+1)
  plt.subplots_adjust( hspace=1 ,wspace=0.5)
  plt.plot(df[price_change[i]],c=color)
  plt.xticks(rotation=90)
  plt.title(price_change[i])
  plt.xlabel('Years')
  plt.ylabel('% Change')
```

the graphs show that plywood has the lowest price change, less than 20, while Soft Sawnwood has the higest price change, moret than 60.
